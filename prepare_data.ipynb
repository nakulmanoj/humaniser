```python
import json
```

```python
with open("data.json", "r", encoding="utf-8") as f:
    data = json.load(f)
formatted=[]
```

```python
cleaned = []
skipped = 0

for item in data:
    neutral = item["input"].strip().replace("\n", " ")
    human = item["output"].strip().replace("\n", " ")
    
    # Skip bad examples
    if len(neutral.split()) < 5:  # Too short
        skipped += 1
        continue
    if len(human.split()) < 3:  # Too short
        skipped += 1
        continue
    if neutral == human:  # No change
        skipped += 1
        continue
    
    text=f"""### Neutral:
{neutral}
### Humanised:
{human}
"""
    cleaned.append({"text": text})

print(f"Original: {len(data)} examples")
print(f"Cleaned: {len(cleaned)} examples")
print(f"Skipped: {skipped} examples")
```

```python
with open("trained.json", "w", encoding="utf-8") as f:
    json.dump(cleaned, f, ensure_ascii=False, indent=2)
```

```python
# Show first 3 examples
print("\nFirst 3 cleaned examples:")
for i in range(min(3, len(cleaned))):
    print(f"\n--- Example {i+1} ---")
    print(cleaned[i]["text"])
```
